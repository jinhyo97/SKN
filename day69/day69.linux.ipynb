# Linux

## Linux 기초

### Operation System (OS)

운영체제란 컴퓨터 하드웨어와 소프트웨어 사이에서 소통을 도와주는 인터페이스. <br>
유닉스, 리눅스, 윈도우 등이 존재. <br>

**유닉스**
교육 및 연구 기관에서 즐겨 사용되는 범용 다중 사용자 방식의 대화식, 시분할처리 시스템용 운영 체제. <br>
1970년대 초반 벨 연구소 직원인 켄 톰슨, 데니스 리치 등이 소형 컴퓨터용으로 개발. <br> 
**유료**

<br>

**리눅스** <br>
유닉스 계열의 운영체제. <br>
핀란드 헬싱키 대학의 리누스 토발즈가 대형 컴퓨터에서 사용하는 유닉스 소스 코드를 수정하여 개인 PC 에서도 사용할 수 있는 **오픈 소스**로 개발.
리눅스는 각종 회사, 커뮤니티에서 만든 배포판이 여러가지 존재. <br>
ex) 우분투, 레드햇, CentOS, 페도라, 데비안 등


### Kernel, Shell

**kernel** <br>
하드웨어와 애플리케이션 간에 상호 작용을 도와주는 OS의 핵심 소프트웨어.

<br>

**shell**

유닉스 계열의 시스템에서 사용하는 대화형 인터페이스. <br>
사용자의 입력을 받아서 명령을 해석하여 커널에 전달하고 결과를 사용자에게 반환. <br>
ex) Bash, Tcsh, Ksh, Zsh, Fish

### Process

실행 중인 프로그램. <br>
여러 개의 프로그램을 동시에 실행하는 것을 멀티태스킹이라고 하고, <br>
이 프로세스를 관리하는 것이 운영체제의 주요 역할 중 하나. <br>
foreground와 background방식으로 동작 <br>

foreground: 사용자의 즉각적인 입력(키보드 입력, 마우스 클릭 등)을 필요로 하는 작업 수행. <br> &nbsp;&nbsp;&nbsp;&nbsp; 실행되면 다른 작업을 처리할 수 없음. <br>
background: 사용자와 입력을 필요로하지 않는 방식. <br>
&nbsp;&nbsp;&nbsp;&nbsp; 프로세스의 실행을 기다리지 않고 다른 작업을 계속 할 수 있음.

> 백그라운드 방식으로 실행해도 사용자의 터미널 세션이 종료되면 실행중인 프로세스도 종료. 
> 작업시간이 오래 걸리는 작업인 경우 nohup 명령어를 이용하여 백그라운드로 실행하면 사용자의 터미널 세션이 종료되어도 작업이 종료될 때까지 프로세스 실행.

<br>

<font style="font-size:18px"> Code </font>

> ```bash
> # 포그라운드 방식 
> $ sample.sh
> 
> # 백그라운드 방식 -- &추가 
> $ sample.sh &
> 
> # nohup 커맨드(no hang up)
> $ nohup sample.sh &
> ```

### Package 설치

apt-get은 데비안 계열에서 사용하는 패키지 설치 명령. <br>
패키지 간 의존성을 확인해서 자동으로 설치. <br>

<br>

|yum|apt-get|
|---|-------|
|CentOS|우분투|

<br>

<font style="font-size:18px"> Command </font>

| 명령                  | 설명                                                     |
|-----------------------|----------------------------------------------------------|
| `apt-get -h`              | 설명 출력                                              |
| `apt-get list [프로그램명]` | 설치할 수 있는 프로그램 버전 출력                    |
| `apt-get install [프로그램명]` | 프로그램 설치                                      |
| `apt-get search [프로그램명]` | 프로그램 검색                                        |
| `apt-get update`          | 설치된 프로그램을 업데이트                               |
| `apt-get update [프로그램명]` | 특정 프로그램을 업데이트                              |
| `apt-get upgrade`         | 설치된 프로그램을 업그레이드 (`update`는 사용하지 않는 패키지를 유지하고, `upgrade`는 삭제) |
| `apt-get remove [프로그램명]` | 프로그램 삭제                                        |

<br>

<font style="font-size:18px"> 주요 옵션 </font>

|옵션	| 설명|
|------|----|
|-y	|모든 옵션에 yes 입력|


<br>

<font style="font-size:18px"> Code </font>

> ```bash
> apt-get install vim
> ```

##### crontab

정기적으로 지정한 시간에 실행하고 싶은 명령어를 등록.

<br>

<font style="font-size:18px"> 주요 옵션 </font>


|옵션	| 설명|
|------|----|
|-l	|등록된 명령어 리스트 확인|
|-e	|등록된 명령어를 수정|
|-r	|등록된 크론탭을 삭제|
|-u	|크론탭을 등록할 사용자 지정|

<br>

<font style="font-size:18px"> Code </font>

> ```bash
> # 크론탭 등록
> $ crontab -e
> 
> # 등록된 크론탭 확인
> $ crontab -l
> ```

<br>

크론탭을 등록할 때는 실행하고자 하는 주기와 명령어를 입력. <br>
주기는 분, 시, 일, 월, 요일의 형태로 입력. <br>
*는 모두 의미. <br>

> ```bash
> # 매 10분에 time.sh 실행
> 10 * * * * time.sh
>
> # 매일 1시에 log 로 끝나는 파일을 찾아서 find.log 파일로 저장
> 0 1 * * * find -name '*.log'
> 
> 
> # 5분 마다 program.sh 실행
> */5 * * * * program.sh
> 
> 
> # 4-10 시 사이에 1시간마다 program.sh 실행
> 0 4-10/1 * * * program.sh
> ```

##### kill

프로세스를 종료. <br>
kill은 프로세스에 시그널을 전송. <br>
시그널을 생략하면 TERM 시그널을 전송하여 프로세스를 종료. <br>
-9 옵션을 이용하여 프로세스를 강제 종료하는 경우도 존재. <br>
-> 프로세스가 종료에 따른 후속처리 없이 종료됩니다. <br>

<br>

<font style="font-size:18px"> 주요 옵션 </font>

|시그널|	번호	|설명|
|----|-----------|---|
|HUP	|1	|재기동|
|INT	|2	|인터럽트|
|QUIT	|3	|종료|
|KILL	|9	|강제종료|
|TERM	|15	|종료|
|STOP	|17	|중단|
|CONT	|19	|재개|

<br>

<font style="font-size:18px"> Code </font>

> ```bash
> # pid 확인
> ps
> 
> # process kill
> $ kill -9 15302
> ```

##### sudo, su

리눅스에서 현재 접속한 계정과 다른 계정에서 작업을 처리해야 할 때 su, sudo를 이용해 처리. <br>
또는 관리자 권한이 필요할 때 사용. <br>

<br>

<font style="font-size:18px"> su </font>

su는 substitute user의 약어. <br>
root 계정이나 다른 계정으로 변경할 때 사용. <br>

> ```bash
> # su를 입력하면 root 계정으로 변경. 암호 필요.
> su
>     
> # test 계정으로 변경 
> su test
> ```

<br>

<font style="font-size:18px"> sudo </font>

su는 substitute user의 약어. <br>
root 계정이나 다른 계정으로 변경할 때 사용. <br>

> ```bash
> # root 계정으로 tmux 설치

> ```

##### systemctl

리눅스 시스템이 가장 먼저 실행되어 다른 프로세스를 실행하는 역할을 하는 것이 systemd(system daemon). <br>
이 프로세스로 프로그램을 서비스로 등록할 때 사용할 수 있는 명령어가 systemctl. <br>

<br>

> ```bash
> # 실행 
> systemctl start zookeeper
> 
> # 정지
> systemctl stop zookeeper
> 
> # 상태 확인 
> systemctl status zookeeper
> 
> # active 상태 확인 
> systemctl is-active zookeeper
> ```

##### chmod

리눅스 파일 사용 권한 설정 <br>

File을 사용해 할 수 있는 작업은 아래의 세 가지로 분류 
- 파일에 저장된 데이터를 읽기. (r = read)
- 파일에 데이터를 쓰기. (w = write)
- 파일 실행. (x = execute)

<br>

<font style="font-size:18px"> 리눅스 파일 권한의 필요성 </font>

모든 파일이 읽기(r), 쓰기(w), 실행(x) 가능하면 보안 및 안정성 문제 발생 가능
- 중요한 파일이 잘못된 명령으로 수정되거나 삭제될 수 있음
- 사용자의 실수로 시스템 전체가 손상될 위험

<br>

<font style="font-size:18px"> 파일 권한의 역할 </font>

각 파일/디렉토리에 대해 읽기(r), 쓰기(w), 실행(x) 권한을 지정 가능. <br>
권한은 아래의 사용자 유형별로 나뉨
- 사용자(user): 파일 소유자.
- 그룹(group): 특정 그룹 소속 사용자.
- 기타 사용자(others): 위 두 유형 외의 모든 사용자.

<br>

<font style="font-size:18px"> Code </font>

>```bash
> chmod u+x FILE                    # 파일 소유 사용자에게 실행권한 추가
> 
> chmod u+w FILE                    # 파일 소유 사용자에게 쓰기 권한 추가
> 
> chmod u=rwx FILE                  # 파일 소유 사용자에게 읽기, 쓰기, 실행 권한 지정
> 
> chmod u-x FILE                    # 파일 소유 사용자의 실행 권한 제거
> 
> chmod g+w FILE                    # 파일 소유 그룹에 쓰기 권한 추가
> 
> chmod g-x FILE                    # 파일 소유 그룹의 실행 권한 제거
> 
> chmod o=r FILE                    # 파일 소유 사용자 및 그룹을 제외한 사용자는 읽기만 가능
> 
> chmod a-x *                       # 현재 디렉토리의 모든 파일에서 모든 사용자의 읽기 권한 제거
> 
> chmod a-w FILE                    # 모든 사용자에 대해 쓰기 권한 제거
> 
> chmod u=rwx,g=r FILE              # 파일 소유 사용자는 모든 권한, 그룹은 읽기만 가능
> 
> chmod ug=rw FILE                  # 파일 소유 사용자와 그룹이 읽기, 쓰기 가능
> 
> chmod g=rw,o=r FILE               # 파일 소유 그룹은 읽기, 쓰기 가능, 그 외 사용자는 읽기만 가능
> 
> chmod ug=rw,o=r FILE              # 파일 소유 사용자 및 그룹은 읽기, 쓰기 가능, 그외 사용자는 읽기만 가능
> 
> chmod 000 FILE                    # 모든 사용자의 모든 권한 제거 = ---------
> 
> chmod 664 FILE                    # 사용자(읽기+쓰기), 그룹(읽기+쓰기), 그외 사용자(읽기) = rw-rw-r--
> 
> chmod 755 FILE                    # 사용자(읽기+쓰기+실행), 그룹(읽기+실행), 그외 사용자(읽기+실행) = rwxr-xr-x
> 
> chmod 777 FILE                    # 모든 사용자에 모든 권한 추가
> 
> chmod -R g+x DIR                  # DIR 디렉토리 하위 모든 파일 및 디렉토리에 그룹 실행(x) 권한 추가
> 
> chmod -R o-wx *                   # 현재 디렉토리의 모든 파일에서 그외 사용자의 쓰기, 실행 권한 제거
> 
> chmod -R a-x,a+X *                # 현재 디렉토리 기준 모든 파일 읽기 권한 제거, 디렉토리 실행 권한 추가
> 
> chmod -R a-x+X *                  # 위(chmod -R a-x,a+X *)와 동일
> 
> chmod u=g FILE                    # FILE의 그룹 권한 값을 사용자 권한으로 적용

>```

##### tar

여러개의 파일을 모아서 tar 형식으로 압축.

<br>

<font style="font-size:18px"> 주요 옵션 </font>

| 옵션          | 내용                                                   |
|---------------|--------------------------------------------------------|
| `-c`          | 파일 묶음 생성                                       |
| `-x`          | 파일 묶음 해제                                       |
| `-z`          | 파일을 gzip 압축하면서 묶음을 생성                       |
| `-v`          | 처리 상황 출력                                       |
| `-t`          | 파일 묶음에 들어있는 파일 목록 출력                   |
| `-P`          | 절대 경로 처리                                         |
| `--exclude`   | 파일 제외. `exclude`는 tar 버전에 따라 커맨드의 제일 앞에 사용되어야 하는 경우도 있음 |
| `--delete`    | 파일 삭제                                              |

<br>

<font style="font-size:18px"> Code </font>

> ```bash
> # 파일 압축
> # foo, bar를 archive.tar로 만듦
> tar -cf archive.tar foo bar
>
> # archive.tar 파일 안에 묶여 있는 내용 확인
> tar -tf archive.tar
>
> # foo, bar를 gzip 압축하여 archive.tar.gz 으로 만듦 
> tar -zcvf archive.tar.gz foo bar
>
> # source_dir의 .log, .attach 로 끝나는 파일, folder_name 디렉토리 제외하고 file.tar.gz 으로 압축 
> tar -zcvf file.tar.gz source_dir --exclude="*.log" --exclude="*.attach*" --exclude="./folder_name"
>
> # source_dir의 logs 폴더는 제외하고 압축. exclude 옵션을 제일 앞에 사용해야 하는 경우가 있음 
> tar --exclude="source_dir/logs" -zcvf file.tar.gz source_dir
>
> # -----------------------------------------------------
>
> # 파일 압축 풀기
> # archive.tar 파일 묶음 해제 
> tar -xf archive.tar
>
> # archive.tar.gz 압축 파일의 묶을 해제 
> tar -zxvf archive.tar.gz
>
> # -----------------------------------------------------
>
> # 파일 삭제
> $ tar --delete --file=txt.tar a.txt
> ```

#### curl

curl은 여러 프로토콜을 이용하여 네트워크 명령을 전송하는 도구. <br>
지원되는 프로토콜 (HTTP, HTTPS, FTP, FTPS, SCP, SFTP, TFTP, DICT, TELNET, LDAP 또는 FILE) 중 하나를 사용하여 서버간에 데이터 전송 가능. <br>
프록시 지원, 사용자 인증, FTP 업로드, HTTP 게시, SSL 연결, 쿠키, 파일 전송 이력서 등과 같은 유용한 기능을 제공.

<br>

<font style="font-size:18px"> 주요 옵션 </font> <br>

|옵션	|내용|
|------|---|
|-o |{파일명}, --ouutput {파일명}	지정한 파일명으로 파일 다운로드|
|-O |{경로}	파일 이름으로 다운로드|
|-s	|추가 정보없이 처리|
|-I	|헤더 정보만 확인|
|-i	|헤더정보 확인|
|-H	|헤더 정보 전달|
|-G	|파라미터 전달|
|-X	|메소드 타입 설정(기본 GET), POST, DELETE 등|
|-d	|전달 데이터, POST 타입일 때 json 형태로 입력하면 데이터 전달|
|-L	|3XX 응답이 오면 경로를 변경|
|-k	|SSL 검증없이 접근. --insecure와 동일|
|-x	|프록시 설정|
|-v	|상세|
|-d/--data-binary	|텍스트를 그대로 전송|
|-T/--upload-file	|파일을 전송|
|--trace-ascii <파일명>	|curl로 송신한 정보, 수신한 정보를 파일로 저장|
|--retry	|재시도 횟수|
|--retry-max-time	|최대 재시도 시간|

## Bash

#### Pipe, Redirection


<font style="font-size:18px"> 파이프 (|) </font>

한 명령(command1)의 표준 출력을 다른 명령(command2)의 표준 입력으로 전달.

<br>

> ```bash
> command1 | command2 # 표준 출력만 전달.
> command1 |& command2 # 표준 출력과 표준 에러 모두 전달.
> ```

<br>

<font style="font-size:18px"> 리다이렉션 </font>

파일이나 명령의 입력/출력을 재지정

|리다이렉션|내용|
|-------|---|
|< | 파일 읽기|
|> | 파일 쓰기 (덮어쓰기)|
|>> | 파일 쓰기 (추가)|
| & | 표준 출력과 표준 에러를 파일로 저장 |
| >>& | 표준 출력과 에러를 파일 끝에 추가 |

<br>

> ```bash
> # 파일 쓰기 (덮어쓰기)
> command > outfile
> 
> # 파일 쓰기 (추가)
> command >> outfile
> ```

#### 환경변수 설정

env: 전역 변수 설정 및 조회 <br>
set: 사용자 환경 변수 설정 및 조회 <br>
export: 사용자 환경 변수를 전역 변수로 설정 <br>
&nbsp;&nbsp;&nbsp;&nbsp;환경 변수 설정은 현재의 세션에만 유효

<br>

<font style="font-size:18px"> Code </font>

> ```bash
> # PARAM을 선언하고 전역 변수(env)와 사용자 환경 변수(set)에서 확인.
> # 사용자 환경 변수(set)에만 값이 존재 
> PARAM=Hello
> env | grep PARAM
> set | grep PARAM
> PARAM=Hello
> 
> # PARAM을 export로 전역 변수로 변경 
> export PARAM
> # 전역 변수(env)와 사용자 환경 변수(set) 모두 존재 
> env | grep PARAM
> PARAM=Hello
> set | grep PARAM
> PARAM=Hello
> ```

|------|---|
|-a	|생성, 변경되는 변수를 export 함|
|-e	|오류가 발생하면 스크립트 종료|
|-x	|수행하는 명령어를 출력 후 실행|
|-c	|다음의 명령을 실행. ex) bash -c "echo 'A'". bash -c date|
|-o	|옵션 설정|

#### 매개변수

<font style="font-size:18px"> 매개변수 </font>

스크립트 실행 시 사용자나 함수에 전달되는 변수. <br>
전달된 매개변수는 공백으로 구분되어 전달.  <br>

| 타입 | 설명                                     |
|------|------------------------------------------|
| `$#` | 전체 변수의 개수                         |
| `$0` | 파일 이름                                |
| `$1` | 전달된 변수 (순서대로 사용 가능)          |
| `$*` | 전체 매개변수 (공백으로 구분)             |

<br>


<font style="font-size:16px"> Code </font>


> ```bash
> # 스크립트 실행 시 매개변수 전달:
> script.sh a b c
> 
> # script.sh에서 해당 매개변수를 전달받을 때는 다음과 같이 사용
> #!/bin/bash
> 
> ## 스크립트 이름 - script.sh
> file_name=$0      
> ## 전체 파라미터 - a b c
> all_param=$*
> ## 파라미터 길이 - 3
> param_length=$#
> 
> ## 공백으로 구분된 파라미터
> param1=$1  # a
> param2=$2  # b
> param3=$3  # c
> ```

|$!	|이전에 실행한 명령어의 PID|

<br>

<font style="font-size:16px"> Code </font>

> echo $$
>
> # 마지막 실행의 종료 인자 
> # 결과가 0이면 이전 실행한 프로세스가 정상 종료이지만, 다른 값이면 오류 발생
> echo $?
>
> # 이전 실행 명령어의 PID
> # nohup 으로 실행한 이전 명령어의 PID를 기록해 둘 수 있음

# tmux

하나의 터미널 세션 안에서 여러 개의 창을 만들고 관리할 수 있게 해주는 도구. <br>
하나의 터미널에서 여러 작업을 동시에 진행하거나, 세션을 백그라운드에서 실행하고 나중에 재접속 가능. <br>

<br>

<font style="font-size:18px"> Code </font> <br>

> ```bash
> tmux                          # session 시작
> tmux new -s [session_name]    # session_name을 갖는 session 시작
> tmux ls                       # session list
> tmux attach -t [session_name] # session 재접속
> 
> ctrl+b, d                     # 현재 session 빠져나오기
> ctrl+b, p                     # 이전 윈도우로 이동
> ctrl+b, n                     # 다음 윈도우로 이동
>
> ctrl+b, "                     # 현재 창을 가로로 나누기
> ctrl+b, %                     # 현재 창을 세로로 나누기
> ctrl+b, 방향키                  # 나뉜 창에서 이동
> ctrl+d                        # pane 및 session 삭제
> ```
