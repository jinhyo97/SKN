# HTML

HTML(HyperText Markup Language)란 웹 페이지를 구성하는데 사용되는 마크업 언어. <br> 웹 브라우저에게 웹 페이지의 구조와 내용을 알려주는 역할. <br>
HTML은 웹 페이지를 구성하는 가장 기본적인 요소 중 하나이며, 웹 개발의 기초 지식으로 여겨짐.

여러 개의 태그로 이루어져 있으며, 이러한 태그를 사용하여 텍스트, 이미지, 비디오 등 다양한 요소를 웹 페이지에 삽입. <br>
태그는 \<태그이름\> 과 같이 꺾쇠 괄호로 묶인 형식으로 작성하며, 종료 태그를 사용하여 태그의 범위 지정. <br>
ex) \<font\>이것은 문단입니다.\</font\>

<br>

<font style="font-size:18px"> Code </font>

> ```html
> <!DOCTYPE html>
> <html>
>   <head>
>     <!-- tab에 표기될 text -->
>     <title>Hello, world!</title>
>   </head>
>   <body>
>     <p>Hello, world!</p>
>     <hr>
>     <img src="https://d3kxs6kpbh59hp.cloudfront.net/community/COMMUNITY/451e60229e7945a5a81258ebb8cabdaa/aaaa744cf80b4c71b2473334addcde9e_1535190560.jpg" width="500">
>   </body>
> </html>

> ```

<br>

<font style="font-size:18px"> attribute </font>
- tag에 추가 정보를 제공
- 시작 태그에 추가하며, 속성 이름과 속성 값으로 구성 <br>
  ex) \<tag [속성 이름]="[속성 값]"> \</tag\>

<br>

<font style="font-size:18px"> comment </font>
- 브라우저에서 무시되며, HTML 코드의 설명이나 메모를 작성할 때 사용. <br>
  \<!-- comment --> 와 같이 사용

## 태그 구조

`기본 tag`
  - \<!DOCTYPE html\> : HTML 문서의 타입을 선언
  - \<html\> : HTML 문서의 시작을 선언
  - \<head\> : tab에 표기할 text, 문서의 메타 데이터를 정의
  - \<title\> : 문서의 제목을 나타내는 태그
  - \<body\> : 문서의 본문이 담겨 있는 부분

<br>

`시작 tag와 종료 tag` <br>
  ex) \<tag\> [text] \</tag\>

<br>

`종료 tag가 없는 empty tag` <br>
  일반적으로 HTML 페이지에서 다른 콘텐츠를 불러오거나 레이아웃을 조정하는 데 사용 <br>
  ex) \<br\>, \<hr\>, \<img\>, \<input\>, \<link\> 등


### head

HTML 문서의 메타 데이터 정의. <br>
웹 브라우저는 메타 데이터인 \<head\> 태그 안에 있는 정보를 브라우저 창에 표시하지 않고, 문서의 정보를 담고 있는 데이터로 처리. <br>
\<head\> 안에는 웹 페이지의 제목(title), 문자 인코딩 방식(meta charset), 스타일 시트(CSS), JavaScript 파일 등의 정보 포함 가능.

<br>

<font style="font-size:18px"> Code </font>

> ```html
> <!DOCTYPE html>
> <html lang="ko">
> <head>
>   <meta charset="UTF-8">
>   <title>문서 제목</title>
>   <link rel="stylesheet" href="style.css">
> </head>
> ```

<br>

`<meta charset="UTF-8">`
- HTML 문서의 문자 인코딩 방식을 **UTF-8**로 설정
- UTF-8은 모든 유니코드 문자를 표현할 수 있어 다국어 웹페이지 제작에 유용

<br>

`<title>`
- 문서의 제목을 정의
- 브라우저의 **제목 표시줄**과 **즐겨찾기 목록**에 표시
- 문서의 주요 내용을 간략히 요약해 표현

<br>

`<link>`
- 외부 리소스를 HTML 문서에 연결
  - `rel`: 리소스와 문서 간의 관계를 정의
    ex) `stylesheet`는 stylesheet와의 연결
  - `href`: 연결할 리소스의 위치 지정.  
    ex) `style.css`는 stylesheet 파일 로드
- ex) `<link rel="stylesheet" href="style.css">`  
  현재 문서에 `style.css` 파일 적용

<br>

`<html lang="ko">`
- HTML 문서의 **언어 속성**을 설정
- `lang="ko"`는 해당 문서가 **한국어**로 작성되었음을 나타냄
- 검색 엔진이 페이지를 분석하거나 웹 접근성을 개선하는 데 중요한 역할

### body

HTML 문서에서 실제로 화면에 나타내는 내용을 담는 tag. <br>
\<body\> tag 안에는 텍스트, 이미지, 비디오, 오디오 등 다양한 미디어 요소를 삽입 가능. <br>
웹 페이지의 레이아웃을 구성하는 다양한 HTML 태그도 포함 가능. <br>
&nbsp;&nbsp;&nbsp;&nbsp; \<header\>, \<nav\>, \<section\>, \<article\>, \<footer\> 등 <br>
이러한 태그들을 조합하여 웹 페이지의 구조와 디자인을 구성 가능. <br>

### HTML 요소

각 요소마다 속성과 사용 방법이 다르므로 웹 개발 시 적절하게 선택하여 사용. <br>

<br>

`텍스트 요소 (Text Semantics)` <br>
특정 텍스트 유형을 정의하는 요소 <br>
&nbsp;&nbsp;&nbsp;&nbsp;  ex) \<em\>, \<strong\>, \<cite\>, \<q\>, \<blockquote\>, \<abbr\>, \<del\>, \<ins\>, \<sup\>, \<sub\>

<br>

`입력 요소 (Form Elements)` <br>
사용자 입력을 받는 요소. <br>
&nbsp;&nbsp;&nbsp;&nbsp;  ex) \<form\>, \<input\>, \<textarea\>, \<label\>, \<select\>, \<option\>, \<button\>

<br>

`멀티미디어 요소 (Multimedia Elements)` <br>
오디오, 비디오 등 멀티미디어를 포함하는 요소. <br>
&nbsp;&nbsp;&nbsp;&nbsp;  ex) \<img\>, \<audio\>, \<video\>

<br>

`블록 요소 (Block-level Elements)` <br>
화면 전체를 채우는 블록 형태의 요소. <br>
일반적으로 새로운 줄에서 시작. <br>
&nbsp;&nbsp;&nbsp;&nbsp;  ex) \<div\>, \<h1\> ~ \<h6\>, \<p\>, \<ul\>, \<ol\>, \<li\>, \<table\>, \<form\>

<br>

`인라인 요소 (Inline Elements)` <br>
블록 요소 내에 포함. <br>
컨텐츠 크기만큼의 영역 차지. <br>
&nbsp;&nbsp;&nbsp;&nbsp;  ex) \<span\>, \<a\>, \<strong\>, \<em\>, \<img\>, \<input\>, \<button\>

<br>

`그룹 요소 (Grouping Content)` <br>
콘텐츠를 묶어서 사용하는 요소. <br>
&nbsp;&nbsp;&nbsp;&nbsp;  ex) \<div\>, \<span\>, \<section\>, \<article\>, \<nav\>, \<aside\>, \<header\>, \<footer\>

#### 텍스트 요소

**텍스트 요소**

- **제목**: `<h1>` ~ `<h6>`  
  문서의 제목을 나타냄. <br>
  숫자가 클수록 중요도가 낮음.

<br>

- **문단**: `<p>`  
  문단을 생성하는 요소.

<br>

- **강조**
  - `<strong>`: **굵은 글씨**로 텍스트를 강조하며, 중요한 내용을 나타냄.
  - `<em>`: *이탤릭체*로 텍스트를 강조하며, 문맥적으로 중요한 부분을 의미.

<br>

- **스타일링**
  - `<b>`: 텍스트를 단순히 **굵게** 표시.
  - `<i>`: 텍스트를 단순히 *이탤릭체*로 표시.
  - `<u>`: 텍스트에 **밑줄**을 표시.

<br>

- **선**
  - `<s>`: 텍스트에 **취소선**을 표시하며, 더 이상 유효하지 않음을 나타냄.
  - `<del>`: **삭제선**을 표시하며, 삭제된 내용을 의미.

<br>

- **기타 강조**
  - `<mark>`: 텍스트를 **하이라이트**하여 강조.
  - `<ins>`: **추가된 내용**임을 나타내며, 밑줄로 표시.

<br>

- **첨자**
  - `<sub>`: **아래 첨자** ex) X₂.
  - `<sup>`: **위 첨자** ex) X².

<br>

- **인용**
  - `<q>`: 짧은 인용문을 생성하며, 따옴표(`"`)로 표시.
  - `<blockquote>`: 긴 인용문을 나타냄.

<br>

- **축약어**
  - `<abbr>`: 축약어를 정의하며, `title` 속성을 이용해 전체 뜻을 제공 가능.  
    ex) `<abbr title="World Health Organization">WHO</abbr>` → WHO  

<br>

**시각적 요소와 의미론적 요소의 차이**
- `<b>`, `<i>`, `<u>`: 단순히 시각적인 스타일을 제공하며, 의미적인 강조 없음.
- `<strong>`, `<em>`, `<ins>`: 문맥상 중요한 내용을 강조하며, **스크린 리더**에서도 강조.
- `<del>`: 삭제된 내용을 나타내며, **문서 수정 이력**을 기록하는 데 유용.
- `<s>`: 더 이상 유효하지 않은 텍스트를 나타내며, 할인 전 가격 같은 경우에 사용.

<br>

<font style="font-size: 18px"> Code </font>

> ```html
> <p>우리는 <strong>필수적으로</strong> 이를 알아야 합니다.</p>
> <p>이 단어는 <b>굵게</b> 표시</p>
> <p><em>이 부분</em>은 강조되어 표시</p>
> <p>이 텍스트는 <s>취소선</s>이 그어짐</p>
> <p><abbr title="World Health Organization">WHO</abbr>은 1948년 설립.</p>
> ```

#### 입력 요소

사용자 입력을 받아 들일 수 있도록 준비가 되어 있으며, 이를 서버로 전송하거나 클라이언트 측 스크립트로 처리 가능. <br>
입력 요소에는 다양한 스타일링을 적용할 수 있어서 웹 페이지에서 다양한 디자인을 구현 가능. <br>

##### input

**공통 속성**
- `name`: 데이터를 서버로 전송할 때 사용할 이름을 지정.
- `value`: 입력 필드의 초기값 설정.
- `placeholder`: 입력 필드에 힌트를 제공하는 문구 설정.
- `required`: 필수 입력 필드 여부 지정.
- `readonly`: 읽기 전용 여부 지정.
- `disabled`: 비활성화 여부 지정.
- `maxlength`: 입력 가능한 최대 문자 수 지정.

<br>

**기타 유용한 속성**
- `autocomplete`: 자동완성 기능 설정.
- `autofocus`: 페이지 로드 시 입력 필드에 자동 포커스 설정.
- `min`, `max`, `step`: 숫자나 날짜 필드의 범위 및 간격 설정.
- `pattern`: 정규 표현식을 이용해 입력 형식 지정.
- `size`: 입력 필드의 가로 크기 설정.

<br>

**입력 타입별 설명 및 예제**

- **텍스트 입력**  
  일반 텍스트를 입력받는 필드
  
  > ```html
  > <input type="text" name="username" placeholder="사용자 이름">
  > ```

<br>

- **비밀번호 입력**
  비밀번호 입력 필드 (입력 내용 숨김 처리)

  > ```html
  > <input type="password" name="password" placeholder="비밀번호">
  > ```

<br>

- **체크박스**
  여러 옵션 중 선택 가능

  > ```html
  > <input type="checkbox" name="hobby" value="music"> 음악
  > <input type="checkbox" name="hobby" value="reading"> 독서
  > ```

<br>

- **라디오 버튼**
  여러 옵션 중 하나만 선택 가능

  > ```html
  > <input type="radio" name="gender" value="male"> 남성
  > <input type="radio" name="gender" value="female"> 여성
  > ```
  
<br>

- **파일 업로드**
  파일 선택 필드

  > ```html
  > <input type="file" name="file" accept=".jpg, .jpeg, .png" multiple>
  > ```
  
<br>

- **재설정 버튼**
  입력 양식 초기화
  
  > ```html
  > <input type="reset" value="재설정">
  > ```

<br>
<br>

<font style="font-size:18px"> **`<input>`과 `<button>` 태그 비교** </font>

<br>

<font style="font-size:18px"> 공통점 </font>

- **공통 `type` 속성값**
  - `reset`: form 태그 내에서 입력 값을 초기화
  - `submit`: form 데이터를 서버로 전송
  - `button`: JavaScript를 통해 사용자 정의 액션 수행

<br>

- **사용 목적**
  - 사용자가 클릭을 통해 특정 동작(초기화, 제출, 커스텀 액션 등)을 수행하도록 설계.


<font style="font-size:18px"> 차이점 </font>

| **특징**               | **`<input>`**                                      | **`<button>`**                                     |
|-----------------------|---------------------------------------------------|--------------------------------------------------|
| **기본 타입**         | `type="text"`                                     | `type="submit"`                                  |
| **디자인**            | 단순한 버튼 구현에 적합. 라벨은 `value` 속성으로 설정 | 복잡한 디자인 가능. 텍스트, 이미지 등 자식 요소 포함 가능 |
| **유연성**            | 제한적 (자식 요소 불가)                              | 높은 유연성 (텍스트나 HTML 요소 포함 가능)            |
| **기능 구현 방식**     | 주로 단순 버튼, 폼 초기화/제출에 사용                | 자식 요소를 통해 유연한 스타일링 및 복잡한 액션 가능.      |
| **명시적 `type` 필요성** | 기본값 `type="text"`, 명시적 설정 필요.               | 기본값 `type="submit"`, 명시적 설정 불필요           |

<br>

<font style="font-size:18px"> Code </font>

`<input>` 태그로 버튼 구현

> ```html
> <input type="button" value="클릭">
> <input type="submit" value="제출">
> <input type="reset" value="초기화">
> ```

<br>

`<button>` 태그로 버튼 구현

> ```html
> <button type="button">클릭</button>
>
> <button type="submit">
>   <img src="submit-icon.png" alt="제출 아이콘">
>   제출
> </button>
> ```


##### form

사용자로부터 입력을 받기 위한 폼을 생성하는데 사용. <br>

<br>


<font style="font-size:18px"> 주요 속성 </font>

- action: 서버로 전송될 데이터의 URL 주소를 지정.
- method: 데이터 전송 방식 지정. (GET, POST)
- target: 서버 응답 결과 표시 위치를 지정.
  - _self: 현재 창 (기본값)
  - _blank: 새 창
  - _parent: 부모 프레임
  - _top: 최상위 창
- enctype: 데이터 인코딩 방식 지정 (주로 파일 업로드에 사용).
  - application/x-www-form-urlencoded (기본값)
  - multipart/form-data: 파일 업로드에 사용.
  - text/plain: 텍스트로 전송.

<br>

<font style="font-size:18px"> 내부 요소 </font>

- \<input\>: 텍스트, 숫자, 체크박스, 라디오 등 다양한 입력 요소 생성.
- \<textarea\>: 여러 줄의 텍스트 입력 영역 생성.
- \<select\>: 드롭다운 메뉴 생성.
- \<button\>: 버튼 생성.

<br>


<font style="font-size:18px"> Code </font>

> ```html
> <!--
> <form action="#" method="post">
>   <label for="name">이름:</label>
>   <input type="text" id="name" name="name"><br>
> 
>   <label for="email">이메일:</label>
>   <input type="email" id="email" name="email"><br>
> 
>   <label for="message">메시지:</label>
>   <textarea id="message" name="message"></textarea><br>
> 
>   <input type="submit" value="제출하기">
> </form>
> -->
> ```

- action="#": 현재 페이지로 폼 데이터를 전송
- method="post": POST 방식으로 데이터 전송
- \<label\>: 폼 요소에 라벨을 추가
  - for: 연결된 폼 요소의 id와 일치해야 함
- \<input\>: 이름과 이메일 입력 필드
- \<textarea\>: 메시지 입력 필드
- \<input type="submit"\>: 제출 버튼

##### textarea

여러 줄의 텍스트를 입력할 수 있는 입력 필드 생성. <br>
사용자로부터 긴 텍스트 입력을 받는 경우에 유용. <br>

<br>

<font style="font-size:18px"> 주요 속성 </font>

- rows: 텍스트 입력란의 표시할 행 수를 지정
- cols: 텍스트 입력란의 표시할 열 수를 지정
- name: 텍스트 입력란의 이름을 지정. 서버로 전송될 때 이 이름을 기반으로 데이터가 전송.
- placeholder: 사용자에게 입력 예시 또는 안내 문구를 제공하는 텍스트를 지정.
- readonly: 읽기 전용으로 텍스트 입력란을 설정. 사용자는 텍스트를 수정할 수 없음.
- disabled: 비활성화된 텍스트 입력란을 설정. 사용자는 텍스트를 입력하거나 수정할 수 없음.

<br>

<font style="font-size:18px"> Code </font>

> ```html
> <textarea id="text_area" name="text_area" cols="30" rows="10" placeholder="Enter your message here"></textarea>
> ```

cols와 rows 속성으로 입력 필드의 크기를 설정 <br> 
placeholder 속성으로 입력 필드에 기본 텍스트를 지정. <br>
입력된 텍스트는 name 속성의 값인 text_area로 서버로 전송. <br>

#### 멀티미디어 요소

이미지, 음성 등의 멀티미디어를 삽입하는 tag. <br>
주요 멀티미디어 요소는 다음과 같음. <br>

- `<img>` : 이미지를 삽입할 때 사용하는 요소. <br>
src 속성으로 이미지 파일의 경로를 지정하며, alt 속성은 이미지가 로드되지 않았을 때 대체 텍스트를 제공함. <br>
- `<audio>` : 음악 파일을 재생할 때 사용하는 요소. <br> src 속성으로 오디오 파일의 경로를 지정하며, controls 속성을 사용하여 오디오 플레이어를 표시할 수 있음.  <br>
- `<video>` : 동영상 파일을 재생할 때 사용하는 요소. <br> src 속성으로 비디오 파일의 경로를 지정하며, controls 속성을 사용하여 비디오 플레이어를 표시할 수 있음. <br>
- `<source>` : `<audio>`나 `<video>` 요소와 함께 사용하여 브라우저가 지원하는 형식의 미디어 콘텐츠를 제공할 수 있음. <br>
- `<canvas>` : 그래픽을 그리기 위한 요소. <br> JavaScript를 사용하여 캔버스에 그림을 그릴 수 있음. <br>
- `<svg>` : 벡터 그래픽을 표시하기 위한 요소. <br> XML 기반의 문서로 작성되며, JavaScript를 사용하여 SVG를 조작할 수 있음. <br>
- `<iframe>` : 다른 웹 페이지를 삽입하기 위한 요소. <br> src 속성으로 웹 페이지의 경로를 지정할 수 있으며, width와 height 속성을 사용하여 크기를 조절할 수 있음. <br>

이 외에도 다양한 멀티미디어 요소가 있으며, 이러한 요소들을 적절히 활용하여 풍부하고 다양한 콘텐츠를 제공할 수 있음.

#### 블록 요소

##### h, p, pre, code

<font style="font-size=18px"> `<h>` </font>

웹 페이지에서 제목을 표시하는 데 사용. <br>
`<h1>`은 가장 중요한 제목을, `<h6>`는 가장 덜 중요한 제목을 나타냄. <br>
제목 태그는 기본적으로 텍스트를 굵고 크게 표시하며, 페이지의 구조를 명확히 하여 검색 엔진 최적화(SEO)에도 유리. <br>
페이지에는 하나의 `<h1>` 태그만 사용하여 주요 제목을 나타냄. <br>

<br>

<font style="font-size=18px"> Code </font>

> ```html
> <h1>웹 페이지 제목</h1>  <!-- 가장 중요한 제목 -->
> <h2>부제목</h2>          <!-- 두 번째로 중요한 제목 -->
> <h3>세부 제목</h3>       <!-- 세 번째로 중요한 제목 -->
> ```

<br>
<br>


<font style="font-size=18px"> `<p>` </font>

텍스트를 하나의 단락으로 묶을 때 사용. <br>
텍스트를 구분하여 읽기 쉽게 만들어 주며, 브라우저에서 자동으로 단락 간 여백을 추가. <br>

<br>

<font style="font-size=18px"> Code </font>

> ```html
> <p> 하나의 단락. 여러 문장이 포함될 수 있음.</p>
> ```

<br>
<br>

<font style="font-size=18px"> `<pre>` </font>

서식이 유지된 텍스트. <br>
주로 코드나 고정 폭 글꼴을 표시할 때 사용. <br>
이 태그 안에 작성된 텍스트는 공백, 줄 바꿈, 탭 등의 서식이 그대로 유지. <br>
-> 코드 블록이나 형식이 중요한 텍스트에 적합.

<br>

<font style="font-size=18px"> Code </font>

> ```html
> <pre>
> function greet() {
>     console.log("Hello, world!");
> }
> </pre>
> ```

<br>
<br>

<font style="font-size=18px"> `<code>` </font>

코드 조각, 변수, 함수 이름, 파일 경로 등을 인라인으로 표시할 때 사용. <br>
코드에 사용되는 글꼴로 스타일링되며, 주로 텍스트 내에서 작은 코드 부분을 강조할 때 사용. <br>

<br>

<font style="font-size=18px"> Code </font>

> ```html
> <p> 변수 <code>varable</code>을 선언.</p>
> ```

##### div

문서의 구조를 분할하거나 구획을 나누는 데 사용. <br>
주로 스타일링이나 레이아웃을 적용할 때 유용한 컨테이너 역할. <br>
-> 특정 스타일을 적용하거나 여러 요소를 그룹화할 때 사용. <br>
CSS와 결합하여 레이아웃, 배경 이미지, 마진, 패딩, 폰트 크기 등 다양한 스타일을 적용 가능. <br>


<font style="font-size=18px"> Code </font>

> ```html
> <div class="header">
>   <h1>웹 페이지 제목</h1>
> </div>
> 
> <div class="content">
>   <p>본문</p>
> </div>
> 
> <div class="footer">
>   <p>웹 페이지 하단 정보</p>
> </div>
> 
> <!-- css 결합-->
> <style>
>   .header {
>     background-color: #f4f4f4;
>     padding: 20px;
>     text-align: center;
>   }
> 
>   .content {
>     padding: 20px;
>     font-size: 16px;
>   }
> 
>   .footer {
>     background-color: #333;
>     color: white;
>     text-align: center;
>     padding: 10px;
>   }
> </style>
> ```

##### ul, ol, li


<font style="font-size:18px"> `<ul>`: Unordered List </font>

순서가 없는 목록을 나타내며, 기본적으로 항목 앞에 Bullet Point 표시.

> ```html

>   <li>항목 3</li>

<br>

<font style="font-size:18px"> `<ol>`: Ordered List </font>

순서가 있는 목록을 나타내며, 기본적으로 숫자가 항목 앞에 표시. <br>

> <ol>
>   <li>첫 번째</li>
>   <li>두 번째</li>
>   <li>세 번째</li>

<font style="font-size:18px"> `<li>`: List Item </font>

목록의 각 항목을 정의하는 태그로, `<ul>` 또는 `<ol>` 태그 내부에서 사용.

##### table

HTML 문서에서 표를 생성할 때 사용되며, 표의 행, 열, 헤더 등을 정의하는 여러 태그 포함.

> ```html
> <table>
>   <tr>
>     <th>이름</th>
>     <th>나이</th>
>     <th>성별</th>
>   </tr>
>   <tr>
>     <td>홍길동</td>
>     <td>30</td>
>     <td>남자</td>
>   </tr>
> </table>
>
> <style>
> table, th, td {
>   border: 1px solid black; /* 테두리 설정 */
>   border-collapse: collapse; /* 테두리 겹침 해제 */
>   text-align: center; /* 가운데 정렬 설정 */
>   padding: 10px; /* 셀 안의 여백 설정 */
> }
> </style>
> ```

- tr: 행
- th: 셀의 헤더
- td: 테이블 셀

    > ```html
    > <td rowspan="2"> <!-- rowspan="2" 속성: 이 셀은 아래 두 행을 합쳐서 하나의 셀로 표현 -->
    > <td colspan="2"> <!-- colspan="2" 속성: 이 셀은 오른쪽 두 열을 합쳐서 하나의 셀로 표현 -->
    > ```

#### 인라인 요소

##### span

텍스트나 구문을 꾸밀 때 사용하는 tag. <br>
\<span\> 태그는 CSS와 함께 사용되어 텍스트나 구문의 일부분에 스타일을 적용할 때 유용. <br> ex) 특정 단어나 구문을 강조하거나, 하이라이트를 적용하는 등의 작업에 사용 가능. <br>

id는 HTML 요소에 고유한 식별자를 지정할 때 사용. <br>
id 속성을 사용하면 CSS나 JavaScript에서 해당 요소를 찾아 조작 가능. <br>

<br>

<font style="font-size:18px"> Code </font>

> ```html
> 나는 <span style="color: blue;">파란색</span>을 좋아합니다.
> This text is <em>emphasized</em> and this text is <strong>strong</strong>.
> <p>오늘의 날씨는 <span id="weather">맑음</span>입니다.</p>
> ```

##### a

`<a>` 태그는 anchor의 약자로 웹 페이지 내 특정 위치로 이동하거나 다른 페이지로 연결할 수 있는 하이퍼링크를 생성하는 데 사용함.

<br>

<font style="font-size:18px"> 주요 속성 </font>
- href: 이동할 URL을 지정하는 필수 속성.
- target: 링크를 클릭했을 때 새 창이나 현재 창에서 링크를 열지 여부를 결정함.
    - _self: 링크를 현재 창에서 열음. (기본값)
    - _blank: 링크를 새 창에서 열음.
- rel: 현재 문서와 링크된 문서 간의 관계를 정의. <br>
        SEO에서 중요한 역할. <br>
        rel="nofollow"는 링크된 페이지에 대한 검색 엔진 크롤러의 추적을 방지하는 데 사용. rel="canonical"은 중복된 콘텐츠의 문제를 해결하기 위해 검색 엔진에게 "이 링크된 페이지가 이 문서의 본래 버전입니다"라는 정보를 제공. <br>
        -> 링크된 문서와 현재 문서의 관계를 명확히하고, 검색 엔진에게 정보를 제공하여 웹 페이지의 SEO를 개선하는 데 중요한 역할.

<br>

<font style="font-size:18px"> Code </font>

> ```html
> <a href="https://example.com">예제 링크</a>
> <a href="https://example.com" target="_blank">새 창에서 열기</a>
> <a href="https://example.com" target="_blank" rel="noopener noreferrer">보안 링크</a> <!-- rel="noopener noreferrer" 속성을 사용하여 보안상의 이유로 새 탭에서 링크를 열 때 원래 페이지와 분리. -->
> ```

##### img

`<img>` 태그는 웹 페이지에 이미지를 삽입하는 데 사용되는 HTML 태그로 시각적 콘텐츠 제공. <br> 이 태그는 src와 alt를 포함해야 함.

<br>

<font style="font-size:18px"> 주요 속성 </font>
- src: 이미지를 가져올 경로(URL)를 지정.
- alt: 이미지가 로드되지 않았을 때 대체 텍스트를 제공. <br>
    - alt 속성을 이용한 이미지 태그의 설명은 검색 엔진에서 중요하게 평가. <br>
    검색 엔진은 이미지 파일 자체를 읽을 수 없기 때문에, alt 속성을 이용해 이미지에 대한 설명을 제공하면 검색 결과에 노출될 확률이 높아짐.
- width: 이미지의 너비를 지정. (픽셀 또는 % 단위)
- height: 이미지의 높이를 지정.
- border: 이미지 주위에 테두리를 설정. (HTML5에서는 사용 권장되지 않음, CSS로 대체)
- align: 이미지와 텍스트의 정렬 방식을 지정. (HTML5에서는 비권장, CSS로 대체)

<br>

<font style="font-size:18px"> Code </font>

> ```html
> <!--
> <img src="image.jpg" alt="example" width="200" height="200" border="1">
> -->
> ```

#### 그룹 요소

하나 이상의 요소를 묶어서 하나의 덩어리로 만드는 역할. <br>
CSS 스타일링이나 자바스크립트 동작을 위한 용도로 사용. <br>
다음과 같은 그룹 요소를 제공. <br>

- `<div>` : 블록 레벨 요소로, 내부에 다른 요소들을 포함 가능.
- `<span>` : 인라인 요소로, 내부에 다른 요소들을 포함 가능.
- `<fieldset>` : 폼 관련 요소로, 관련된 입력 필드를 묶어서 그룹화 가능.
- `<legend>` : `<fieldset>` 요소의 제목을 나타냄.
- `<optgroup>` : `<select>` 요소에서 선택지들을 그룹화할 때 사용함.

<br>

그룹 요소는 화면에 나타나는 영역을 정의하지 않기 때문에, 주로 CSS 스타일링을 통해 레이아웃을 구성하는 데 활용. <br>
&nbsp;&nbsp;&nbsp;&nbsp; ex) `<div>` 요소를 사용하여 특정 영역을 감싸서 배경 색상이나 테두리 등의 스타일 지정 가능. <br>
자바스크립트에서 그룹 요소를 사용하여 특정 요소들을 묶어서 처리 가능. <br>

##### select

드롭다운 생성. <br>

<br>

<font style="font-size:18px"> Code </font>

> ```html
> <label for="colors">Choose a color:</label>
> <select id="colors">
>   <option value="red" selected>Red</option>
>   <option value="green">Green</option>
>   <option value="blue">Blue</option>
> </select>
> ```

- label: 드롭다운 메뉴 설명
- for: select 요소와 label 요소를 연결해주는 역할.
- name: 서버에서 전송되는 데이터의 이름
- id: label 요소와 for 속성을 연결해주는 역할.
    - 위 코드에서 초기 선택값은 태그안의 selected에 의해 Red로 설정

##### fieldset

`<form>`은 입력 양식의 전체적인 컨테이너를 정의하고 서버에 데이터를 전송하는 역할을 함. <br>
`<fieldset>`은 입력 요소들을 관련 있는 그룹으로 묶어서 시각적으로 구분하고, 레이블을 제공하는 용도로 사용. <br>
따라서 `<fieldset>`은 일반적으로 `<form>` 안에서 사용되며, 다른 요소들과 함께 다양한 방식으로 조합하여 사용 가능. <br>
    ex) `<legend>` 태그를 사용하여 `<fieldset>`에 대한 제목을 지정하거나, CSS 스타일을 적용하여 시각적인 디자인 구성 가능.

ex) 회원가입 폼을 만든다고 가정. <br>
이 폼에는 이름, 이메일, 비밀번호 등 여러 가지 입력 필드가 있을 수 있음. <br>
이때, 입력 필드들을 모두 감싸고 있는 공통된 영역이 필요하면 이를 `<fieldset>`으로 묶어주는 것이 좋음.

<br>

<font style="font-size:18px"> Code </font>

> ```html
> <!--
> <form>
>   <fieldset>
>     <legend>회원가입 양식</legend>
>     <label for="name">이름:</label>
>     <input type="text" id="name" name="name"><br>
> 
>     <label for="email">이메일:</label>
>     <input type="email" id="email" name="email"><br>
> 
>     <label for="password">비밀번호:</label>
>     <input type="password" id="password" name="password"><br>
> 
>     <input type="submit" value="회원가입">
>   </fieldset>
> </form>
> -->
> ```

#### 의미 요소

요소가 자기 스스로 브라우저와 개발자 모두에게 자신이 사용된 의미를 명확히 전달해 주는 요소. <br>
다음과 같은 semantic tag 제공. <br>

- `<header>` : 웹 페이지나 섹션의 제목이나 로고 등, 상단에 배치되는 콘텐츠를 정의하는 태그. 
- `<nav>` : 웹 페이지에서 주요 링크를 포함하는 태그. <br> 이 태그를 사용하면 해당 영역이 탐색 영역임을 명확하게 표시할 수 있음.
- `<main>` : 웹 페이지의 주요 콘텐츠 영역을 정의하는 태그.
- `<article>` : 독립적으로 구성될 수 있는 콘텐츠 영역을 정의하는 태그. <br>
블로그 글이나 뉴스 기사와 같은 콘텐츠 영역에 사용할 수 있음.
- `<section>` : 웹 페이지의 구분된 섹션을 정의하는 태그. <br>이 태그를 사용하여 웹 페이지의 콘텐츠를 더욱 구조화할 수 있음.
- `<aside>` : 웹 페이지의 주요 콘텐츠와 연관된 보조 정보를 제공하는 태그. <br> 사이드바나 광고 등의 콘텐츠를 이 태그로 정의할 수 있음.
- `<footer>` : 웹 페이지나 섹션의 바닥글을 정의하는 태그. <br> 저작권 정보나 연락처 정보 등을 포함하는 영역으로 사용될 수 있음.
- `<figure>` : 이미지나 삽화, 도표 등과 같은 콘텐츠와 해당 콘텐츠와 관련된 캡션을 포함하는 영역을 정의하는 태그.
- `<figcaption>` : `<figure>` 태그에서 정의한 콘텐츠와 관련된 캡션을 정의하는 태그. <br>
이러한 시맨틱 태그를 사용하면 웹 페이지를 보다 의미 있는 구조로 구성할 수 있으며, 웹 접근성 및 검색 엔진 최적화에도 도움이 됨.
